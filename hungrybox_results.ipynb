```python
#Adattisztítás
def clean_data(rows):
    cleaned_data = []

    for row in rows:
        columns = row.find_all('td')
        if columns:
            data = [col.text.strip() for col in columns]

            # \xa0 -> space
            data = [d.replace('\xa0', ' ') for d in data]

            # Helyezések javítása
            if len(data) > 1:
                data[1] = re.sub(r'^[A-Z]+(\d+)', r'\1', data[1])  # Levágja az előző betűket
                data[1] = re.sub(r'^(\d)(\d+)(st|nd|rd|th)$', r'\2\3', data[1])  # Eltávolítja a felesleges első számjegyet

            cleaned_data.append(data)

    return cleaned_data

# Adatok tisztítása
cleaned_rows = clean_data(rows)

# Ellenőrzés terminálon
for data in cleaned_rows:
    print(data)
```

```python
# CSV fájlba írás
csv_filename = "hungrybox_results.csv"
with open(csv_filename, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)

    # Fejléc
    writer.writerow(["Date", "Place", "Event", "Result", "Prize", "Opponent", "Score"])

    # Sorok beírása
    for data in cleaned_rows:
        writer.writerow(data)

print(f"✅ Az adatok sikeresen kiíródtak a '{csv_filename}' fájlba!")
```
