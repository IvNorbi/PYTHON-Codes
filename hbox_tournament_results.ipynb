```python
# Adattisztítás
def clean_data(rows):
    cleaned_data = []

    for row in rows:
        columns = row.find_all('td')
        if columns:
            data = [col.text.strip() for col in columns]

            #  \xa0 -> space
            data = [d.replace('\xa0', ' ') for d in data]

            # Helyezések előtti karakterek eltávolítása
            data = [d.lstrip('ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz') for d in data]

            cleaned_data.append(data)

    return cleaned_data


cleaned_rows = clean_data(rows)
for data in cleaned_rows:
    print(data)
```

```python
#Adatok exportálása CSV-be

# Fejlécek
headers = ['Date', 'Place', 'Event', 'Result', 'Prize', 'Column6', 'Column7']

# Adatok
data = cleaned_rows  # Ez az, amit a korábbi kódban szereplő tisztított adatok tartalmaznak

with open('hbox_tournament_results.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(headers)
    writer.writerows(data) 
print("Adatok mentve CSV formátumba.")
```
